```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv('Tingkat Setengah Pengangguran Menurut Provinsi, 2015-2017.csv')
df1.head()
```

```python
df2 = pd.read_csv('Tingkat Setengah Pengangguran Menurut Provinsi, 2018-2020.csv')
df2.head()
```

```python
df3 = pd.read_csv('Tingkat Setengah Pengangguran Menurut Provinsi, 2021-2022.csv')
df3.head()
```

```python
df = pd.merge(df1,df2,on='Provinsi')
df = pd.merge(df,df3,on='Provinsi')
df.head(50)
```

```python
df.describe()
```

```python
df.set_index('Provinsi', inplace=True)

# Select the row data for the desired province, e.g., 'ACEH'
row_data = df.loc['INDONESIA']

# Extracting year columns for plotting
x_axis_labels = row_data.index
y_axis_values = row_data.values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_axis_labels, y_axis_values, marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Yearly Data for Indonesia')
plt.grid(False)
plt.show()
```

```python
highest_results = {}
for year in df.columns[1:]:
    highest_province = df.loc[df[year].idxmax(), df.columns[0]]
    highest_rate = df[year].max()
    highest_results[year] = (highest_province, highest_rate)

# Display the result
print("Province with the highest unemployment rate for each year:")
for year, (province, rate) in highest_results.items():
    print(f"{year}: {province} with a rate of {rate:.2f}")
```

```python
lowest_results = {}
for year in df.columns[1:]:
    highest_province = df.loc[df[year].idxmin(), df.columns[0]]
    highest_rate = df[year].min()
    lowest_results[year] = (highest_province, highest_rate)

# Display the result
print("Province with the highest unemployment rate for each year:")
for year, (province, rate) in lowest_results.items():
    print(f"{year}: {province} with a rate of {rate:.2f}")
```

```python
# Prepare data for plotting
years = list(highest_results.keys())
highest_rates = [highest_results[year][1] for year in years]
lowest_rates = [lowest_results[year][1] for year in years]

# Plotting
plt.figure(figsize=(12, 6))

# Plot highest rates
plt.plot(years, highest_rates, marker='o', label='Highest Unemployment Rate', color='red')

# Plot lowest rates
plt.plot(years, lowest_rates, marker='o', label='Lowest Unemployment Rate', color='green')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Highest and Lowest Unemployment Rates by Year')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Print the results for verification
print("Highest unemployment rates:")
for year, (province, rate) in highest_results.items():
    print(f"{year}: {province} with a rate of {rate:.2f}")

print("\nLowest unemployment rates:")
for year, (province, rate) in lowest_results.items():
    print(f"{year}: {province} with a rate of {rate:.2f}")
```
